```python
import matplotlib.pyplot as plt
```

```python
#see plots inside jupyter notebook
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
# Functional method
plt.plot(x,y)
# plt.show() to print out the plot
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
plt.subplot(1,2,1) # row, column, figure you're working on
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
# Object Oriented Method
fig = plt.figure()
# percent left of axes, percent bottom of axes, percentage width, percentage height
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Set Title')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

axes1.plot(x,y)
axes1.set_title('Bigger Plot')
axes2.plot(y,x)
axes2.set_title('Smaller Plot')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2) # essentially calling .figure and adding axes for you
# for current_ax in axes:
#     current_ax.plot(x,y)

# axes is just an array of axes objects
axes[0].plot(x,y)
axes[0].set_title('First Plot')
axes[1].plot(y,x)
axes[1].set_title('Second Plot')

plt.tight_layout() # fix tight overlaps
```

## Figure Size and DPI

```python
fig = plt.figure(figsize=(8,2)) # figsize is in inches
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1,figsize=(8,2)) # figsize is in inches
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

```python
fig.savefig('my_picture.png')
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
# ax.set_title('Title')
# ax.set_ylabel('Y')
# ax.set_xlabel('X')

ax.plot(x,x**2, label='X Squared')
ax.plot(x,x**3, label='X Cubed')
ax.legend(loc=(0.1, 0.1))
```

## Plot Appearance

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color='#FF8C00', linewidth=3, linestyle='--', marker='o', markersize=10)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color='purple', lw=2, marker='o')

# zoom in on the axis
ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
