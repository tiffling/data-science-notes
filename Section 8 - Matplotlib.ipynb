```python
import matplotlib.pyplot as plt
```

```python
#see plots inside jupyter notebook
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
# Functional method
plt.plot(x,y)
# plt.show() to print out the plot
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
plt.subplot(1,2,1) # row, column, figure you're working on
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
# Object Oriented Method
fig = plt.figure()
# percent left of axes, percent bottom of axes, percentage width, percetnage height
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Set Title')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

axes1.plot(x,y)
axes1.set_title('Bigger Plot')
axes2.plot(y,x)
axes2.set_title('Smaller Plot')
```
