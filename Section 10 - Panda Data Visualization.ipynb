```python
import numpy as np
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
df1 = pd.read_csv('Python-for-Data-Visualization/Pandas Built-in Data Viz/df1', index_col=0)
```

```python
df1.head()
```

```python
df2 = pd.read_csv('Python-for-Data-Visualization/Pandas Built-in Data Viz/df2')
```

```python
df2.head()
```

```python
# histograms are bar charts for frequency
df1['A'].hist()
```

```python
# use .plot to make any graph off a dataframe
df1['A'].plot(kind='hist')
```

```python
df1['A'].plot.hist()
```

```python
# like line plot but filled in
df2.plot.area(alpha = 0.4)
```

```python
# takes index as category and maps bars for each col
df2.plot.bar()
```

```python
# all matplotlib options are applicable here
df1.plot.line(x=df1.index, y='B', figsize=(12,3), lw=1)
```

```python
# Third dimension determined by color 'C'
df1.plot.scatter(x='A', y='B', c='C', cmap='coolwarm')
```

```python
# third dimension determined by size 'C'
df1.plot.scatter(x='A', y='B', s=df1['C']*100)
```

```python
df2.plot.box()
```
